```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dates = ['10-03-16', '10-04-16', '10-05-16', '10-06-16', '10-07-16']
opening_prices = [774.25, 776.030029, 779.309998, 779, 779.659973]
max_prices = [776.065002, 778.710022, 782.070007, 780.47998, 779.659973]
min_prices = [769.5, 772.890015, 775.650024, 775.539978, 770.75]
closing_prices = [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]

plt.figure(figsize=(10, 6))
plt.plot(dates, opening_prices, label='Открытие', marker='o')
plt.plot(dates, max_prices, label='Максимум', marker='o')
plt.plot(dates, min_prices, label='Минимум', marker='o')
plt.plot(dates, closing_prices, label='Закрытие', marker='o')

plt.title('Финансовые данные Alphabet Inc. (03.10.2016 - 07.10.2016)')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
people = ('G1','G2','G3','G4','G5','G6','G7','G8')
segments = 4

data = [
    [ 3.40022085,   7.70632498,   6.4097905,   10.51648577,   7.5330039, 7.1123587,   12.77792868,   3.44773477],
    [ 11.24811149,   5.03778215,   6.65808464,  12.32220677,   7.45964195, 6.79685302,   7.24578743,   3.69371847],
    [ 3.94253354,   4.74763549,  11.73529246,   4.6465543,   12.9952182, 4.63832778,  11.16849999,   8.56883433],
    [ 4.24409799,  12.71746612,  11.3772169,    9.00514257,  10.47084185, 10.97567589,   3.98287652,   8.80552122]
    ]


percentages = (np.random.randint(5,20, (len(people), segments)))
y_pos = np.arange(len(people))

fig = plt.figure(figsize=(10,9))
ax = fig.add_subplot(111)

colors =['r', 'b', 'g', 'y']
patch_handles = []

left = np.zeros(len(people))
for i, d in enumerate(data):
    patch_handles.append(ax.barh(y_pos, d, color=colors[i%len(colors)], align='center', left=left))
    left += d

for j in range(len(patch_handles)):
    for i, patch in enumerate(patch_handles[j].get_children()):
        bl = patch.get_xy()
        x = 0.5*patch.get_width() + bl[0]
        y = 0.5*patch.get_height() + bl[1]
        ax.text(x,y, "%d%%" % (percentages[i,j]), ha='center')

ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.set_xlabel('Результаты')
plt.show()
```
